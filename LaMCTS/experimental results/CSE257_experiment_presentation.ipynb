```python
import numpy as np
import matplotlib.pyplot as plt
```

Levy

```python
# Compare BO's performances for Rastrigin function with and that of LA-MCTS
bo = []#to store f(x) found from Rastrigin function 
lamcts = []
nev = []

text_file = open("Levy/bo_result200", "r")
lines = text_file.readlines()
for line in lines:
    tmp = []
    line = line.replace('[', '')
    line = line.replace(']', '')
    arr = line.split(',')
    for i in arr:
        f = float(i)
        tmp.append(f)
    bo = tmp
    
text_file = open("Levy/lamcts_result200", "r")
lines = text_file.readlines()
for line in lines:
    tmp = []
    line = line.replace('[', '')
    line = line.replace(']', '')
    arr = line.split(',')
    for i in arr:
        f = float(i)
        tmp.append(f) 
    lamcts = tmp

text_file = open("Levy/nev_result200", "r")
lines = text_file.readlines()
for line in lines:
    tmp = []
    line = line.replace('[', '')
    line = line.replace(']', '')
    arr = line.split(',')
    for i in arr:
        f = float(i)
        tmp.append(f) 
    nev = tmp

plt.plot(bo)
plt.plot(lamcts)
plt.plot(nev)
plt.legend(['BO', 'LA-MCTS','Nevergrad'])
plt.xlabel("iterations")
plt.ylabel("current best f(x)")
plt.title("The Optimization for the Levy function")
```

Ackley & Gaussian Noise

```python
# Compare LA-MCTS's performances for Ackley function with and without Gaussian Noise(GN)
text_file = open("Ackley/lamcts_d10_iter200", "r")
lines = text_file.readlines()
lamcts = []#to store f(x) found from Ackley function without Gaussian Noise
lamctsWGN = []#to store f(x) found from Ackley function with Gaussian Noise
counter = 0
for line in lines:
    tmp = []
    line = line.replace('[', '')
    line = line.replace(']', '')
    arr = line.split(',')
    for i in arr:
        f = float(i)
        tmp.append(f)
    if counter  == 0:
        lamcts = tmp
    else:
        lamctsWGN = tmp
    counter += 1
```

```python
# Compare BO's performances for Ackley function with and without Gaussian Noise(GN)
text_file = open("Ackley/bo_d10_iter200", "r")
lines = text_file.readlines()
bo = []#to store f(x) found from Ackley function without Gaussian Noise
boWGN = []#to store f(x) found from Ackley function with Gaussian Noise
counter = 0
for line in lines:
    tmp = []
    line = line.replace('[', '')
    line = line.replace(']', '')
    arr = line.split(',')
    for i in arr:
        f = float(i)
        tmp.append(f)
    if counter  == 0:
        bo = tmp
    else:
        boWGN = tmp
    counter += 1
```

```python
# Compare Nevergrad's performances for Ackley function with and without Gaussian Noise(GN)
text_file = open("Ackley/nev_d10_iter200", "r")
lines = text_file.readlines()
nev = []#to store f(x) found from Ackley function without Gaussian Noise
nevWGN = []#to store f(x) found from Ackley function with Gaussian Noise
counter = 0
for line in lines:
    tmp = []
    line = line.replace('[', '')
    line = line.replace(']', '')
    arr = line.split(',')
    for i in arr:
        f = float(i)
        tmp.append(f)
    if counter  == 0:
        nev = tmp
    else:
        nevWGN = tmp
    counter += 1
```

```python
plt.plot(bo)
plt.plot(lamcts)
plt.plot(nev)
plt.legend(['BO','LA-MCTS','Nevergrad'])
plt.xlabel("iterations")
plt.ylabel("current best f(x)")
plt.title("Working on Ackley-10d without Gaussian Noise")
```

```python
plt.plot(boWGN)
plt.plot(lamctsWGN)
plt.plot(nevWGN)
plt.legend(['BO','LA-MCTS','Nevergrad'])
plt.xlabel("iterations")
plt.ylabel("current best f(x)")
plt.title("Working on Ackley-10d with Gaussian Noise")
```
